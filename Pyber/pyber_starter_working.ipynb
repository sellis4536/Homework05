```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Read the City and Ride Data

# Combine the data into a single dataset

# Display the data table for preview
```

```python
ride_df.head()
```

```python
city_df.head()
```

```python
# merge city and ride tables to city_ride_table via left merge

city_ride_table = pd.merge(ride_df, city_df, on="city", how = "left")
city_ride_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
# create city x type df with x, y and s values via groupby w/ .agg

plot_city_ride = city_ride_table.groupby(['city', 'type']).agg({'ride_id': 'count', 'fare': 'mean', 
                                                                'driver_count': 'mean'}).reset_index()
plot_city_ride.head()
```

```python
# First attempts at establishing plot axis

#x = city_ride_table.groupby(['city','type'])['ride_id'].count().reset_index()
#axis=x['ride_id']
#xaxis

#y = city_ride_table.groupby(['city','type'])['fare'].mean().reset_index()
#yaxis=y['fare']
#yaxis

#Urban = city_ride_table['type'] == 'Urban'
#Urban
```

```python
# method 2 - use plt. functions to control line by line
plt.scatter(plot_city_ride['ride_id'], plot_city_ride['fare'], plot_city_ride['driver_count'] *10,)
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Avg. Fare ($)")
plt.show()
```

```python
 # method 1 -- Create the scatter plot w/ s data for bubble plot, control all settings - limited for axis labels so far
    
plot_city_ride.plot.scatter(x='ride_id', y='fare', s=plot_city_ride['driver_count']*10,
                            title = 'Pyber Ride Sharing Data (2016)',
                            alpha = 0.5, grid=True)
plt.show()
```
